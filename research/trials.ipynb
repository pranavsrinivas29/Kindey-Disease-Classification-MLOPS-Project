```python
def prod(x1:int, x2:int)->int:
    return x1*x2
```

```python
prod(2,2)
```

```python
prod(2, "2")
```

### This is because of python's logic which considers higher datatype

```python
from ensure import ensure_annotations

@ensure_annotations
def prod(x1:int, x2:int)->int:
    return x1*x2
```

```python
prod(2, 2)
```

```python
prod(2, "2")
```
